# My prediction
##### The more weight there is the more displacement there is. However if the displacement increase. it will later bloom out and not be linear any more as you can see at the end.

```python
model = LinearRegression()
pol_0 = PolynomialFeatures(degree=3)

model.fit(pol_0.fit_transform(x), y)
x1 = np.linspace(x.min(), x.max(), 10000).reshape(-1,1) 
plt.plot(x, y, 'o')
plt.plot(x1, model.predict(pol_0.fit_transform(x1))) 
plt.show()
```
